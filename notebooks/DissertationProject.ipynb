```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Read in the CSV file in and create a DataFrame
vgs_df = pd.read_csv("data/vgsales.csv", header=0, sep=",")
```

```python
# Presentation in the console or notebook
pd.options.display.max_columns = False
pd.options.display.expand_frame_repr = False
```

```python
# Print out the whole data frame with the print() method
def print_the_whole_data_frame(data_frame):
    print(data_frame)

print_the_whole_data_frame(vgs_df)
```

```python
# Print out the head of the data frame with the head() method
def print_the_head_of_the_data_frame(data_frame, quantity):
    print(data_frame.head(quantity))

print_the_head_of_the_data_frame(vgs_df, 5)
```

```python
# Print out the end of the data frame with the tail() method
def print_the_end_of_the_data_frame(data_frame, quantity):
    print(data_frame.tail(quantity))

print_the_end_of_the_data_frame(vgs_df, 5)
```

```python
# Print out the "info" of the data frame with the info() method
def print_the_info_of_the_data_frame(data_frame):
    data_frame.info()

print_the_info_of_the_data_frame(vgs_df)
```

```python
# Print out the description of the data frame with the describe() method
def print_the_description_of_the_data_frame(data_frame):
    print(data_frame.describe())
    
print_the_description_of_the_data_frame(vgs_df)
```

```python
# Percentile for the Japanese sales
def percentile_test_in_column(data_frame, column, percentile_in):
    sales = data_frame[column]
    percentile_out = np.percentile(sales, percentile_in)
    print(percentile_out)
    
percentile_test_in_column(vgs_df, "JP_Sales", 62)
percentile_test_in_column(vgs_df, "JP_Sales", 63)
```

```python
# Remove empty cells or rows
def remove_empty_cells(data_frame):
    data_frame.dropna(inplace=True)
    
remove_empty_cells(vgs_df)
```

```python
# Delete the columns that are not needed
def delete_the_not_needed_column(data_frame, column):
    del data_frame[column]

delete_the_not_needed_column(vgs_df, "Rank")
delete_the_not_needed_column(vgs_df, "JP_Sales")
delete_the_not_needed_column(vgs_df, "Publisher")
delete_the_not_needed_column(vgs_df, "Year")
```

```python
# Remove the rows of games that has no sales in all the columns
def remove_zero_sales(data_frame, column):
    data_frame.dropna(subset=[column],
                     inplace=True)
    
remove_zero_sales(vgs_df, "NA_Sales")
remove_zero_sales(vgs_df, "EU_Sales")
remove_zero_sales(vgs_df, "Other_Sales")
remove_zero_sales(vgs_df, "Global_Sales")
```

```python
# Delete the rows that did not reach 1 million sales on the global scale
def delete_sales_from_column_below_quantity(data_frame, column, quantity):
    for x in data_frame.index:
        if data_frame.loc[x, column] < quantity:
            data_frame.drop(x, inplace=True)
    return data_frame

delete_sales_from_column_below_quantity(vgs_df, "Global_Sales", 1)
```

```python
# Remove duplicates from the leftover
def remove_duplicates(data_frame):
    data_frame.drop_duplicates(inplace=True)

remove_duplicates(vgs_df)
```

```python
# Print "info" after cleaning
print_the_info_of_the_data_frame(vgs_df)
```
